```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Step 1: Upload an image
uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]
```

```python
# Step 2: Read the image (color)
image = cv2.imread(filename)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for display
```

```python
# Step 3: Apply Mean Filtering (Average filter)
mean_filtered = cv2.blur(image, (5, 5))  # kernel size 5x5
```

```python
# Step 4: Apply Median Filtering
median_filtered = cv2.medianBlur(image, 5)  # kernel size 5
```

```python
# Step 5: Display results
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(mean_filtered)
plt.title("Mean Filtered Image")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(median_filtered)
plt.title("Median Filtered Image")
plt.axis('off')

plt.show()
```
