```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "geeks.png"
image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
```

```python
# Image Sharpening using Laplacian filter
laplacian = cv2.Laplacian(image, cv2.CV_64F)
sharpened = cv2.convertScaleAbs(image - laplacian)
```

```python
# Edge Detection using Gradient filters (Sobel)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # gradient in x-direction
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # gradient in y-direction

```

```python
# Combine gradients
sobel_combined = cv2.magnitude(sobel_x, sobel_y)
```

```python
# Display results
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened Image (Laplacian)")
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(sobel_x, cmap='gray')
plt.title("Sobel X (Vertical Edges)")
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(sobel_y, cmap='gray')
plt.title("Sobel Y (Horizontal Edges)")
plt.axis('off')

plt.show()
```

```python
# Display combined gradient separately
plt.figure(figsize=(6,6))
plt.imshow(sobel_combined, cmap='gray')
plt.title("Edge Detection (Sobel Combined)")
plt.axis('off')
plt.show()
```
