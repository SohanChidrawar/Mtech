```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "geeks.png"
image = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
```

```python
#define intensity slicing range
lower , upper = 100, 200
```

```python
#binary slicing: highlight pixel in range to supress other
binary_sliced = np.where((image >= lower) & (image <= upper),255,0).astype(np.uint8)
```

```python
#range slicied: keep original values in range and suppress other
range_sliced = np.where((image >= lower) & (image <= upper), image, 0).astype(np.uint8)
```

```python
# display the result

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(image,cmap = 'gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(binary_sliced,cmap = 'gray')
plt.title('Binary Image')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(range_sliced,cmap = 'gray')
plt.title('Range Image')
plt.axis('off')

plt.show()
```
